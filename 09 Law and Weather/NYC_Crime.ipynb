Plot Max temp over crime count by day

```python
#Sourced code from: https://matplotlib.org/gallery/api/two_scales.html
t = crimeWeather.date
data1 = crimeWeather.crimeCount
data2 = crimeWeather.TMax

fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Crime Count', color = color)
ax1.plot(t, data1, color = color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Max Temperture', color = color)  
ax2.plot(t, data2, color=color)
ax2.tick_params(axis ='y', labelcolor = color)

fig.tight_layout()  # with out this the right y-label is cut off
plt.title('Crime Count VS. Max Temp by day')
plt.show()
```

```python
import matplotlib.pyplot as pyplot
crimeWeatherSmall = crimeWeather[['date', 'crimeCount', 'TMax', 'Precip']]
values = crimeWeatherSmall.values
# specify columns to plot
groups = [1,2,3]
i = 1
# plot each column
pyplot.figure()
for group in groups:
	pyplot.subplot(len(groups), 1, i)
	pyplot.plot(values[:, group])
	pyplot.title(crimeWeatherSmall.columns[group], y=0.5, loc='right')
	i += 1
pyplot.show()
```

```python
# renaming columns for prophet compatability
ny_crime = crimeWeatherSmall[['date', 'crimeCount']]
ny_crime = ny_crime.rename(columns={'date': 'ds', 'crimeCount': 'y'},index=str)

ny_crime.head()

# Facebook Prophet forecasting library
from fbprophet import Prophet
# create a model using Prophet
m = Prophet()
m.fit(ny_crime)

future = m.make_future_dataframe(periods=365)

forecast = m.predict(future)

fig1 = m.plot(forecast)

```

```python
fig2 = m.plot_components(forecast)
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
Rain = crimeWeather[['date', 'Precip']]
rain_holidays= Rain.loc[(Rain['Precip'] >= 10)] 
```

```python
rain_list = rain_holidays['date'].tolist()
```

```python
holidays = pd.DataFrame({
  'holiday': 'rain',
  'ds': pd.to_datetime(rain_list),
  'lower_window': 0,
  'upper_window': 1,
})
```

```python
m2 = Prophet(holidays=holidays)
forecast2 = m2.fit(ny_crime).predict(future)

#if you run into an error here, try to run the following install:
#!pip install pandas --upgrade
```

```python
# filter forecast results to only days where 'holiday' makes impact
forecast2[(forecast2['holidays']).abs()>0][
    ['ds','holidays']][-10:]
```

```python
fig2 = m2.plot_components(forecast2)
```
